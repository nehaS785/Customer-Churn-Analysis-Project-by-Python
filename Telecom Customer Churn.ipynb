```python
plt.figure(figsize=(4,4))
# Assuming 'df' is your DataFrame
# Step 1: Calculate percentages
data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
percentages = data.div(data.sum(axis=1), axis=0) * 100  # Convert to percentage

# Step 2: Plot the stacked bar chart
percentages.plot(kind='bar', stacked=True, figsize=(6, 4), color=['skyblue', 'orange'])

# Step 3: Add percentage labels on the bars
for i, (index, row) in enumerate(percentages.iterrows()):
    cumulative = 0
    for j, value in enumerate(row):
        plt.text(i, cumulative + value / 2, f"{value:.1f}%", ha='center', va='center', fontsize=9)
        cumulative += value

# Add labels and title
plt.title('Churn by SeniorCitizen (Percentage)', fontsize=14)
plt.xlabel('SeniorCitizen', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```
